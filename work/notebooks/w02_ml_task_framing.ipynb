## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

My lane focuses on identifying web pages that should be refreshed or reviewed for better search performance. This is a classification problem because each page is assigned one of two outcomes:

Needs Refresh (1)
Does Not Need Refresh (0)

The model learns patterns from historical page data and predicts whether a page should be reviewed. This helps prioritize work instead of manually checking every page.


```python
# ML task type
task_type = "Classification"

print("ML Task Type:", task_type)
```

## 3. Success metric

*One metric you can defend. What number means 'good'?*

The primary success metric is Precision@50.

This measures how many of the top 50 pages recommended by the model actually need refreshing.

A high Precision@50 means the model helps content teams focus on the most valuable pages first while reducing unnecessary reviews.

```python
metric = "Precision@50"

print("Success Metric:", metric)
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

A fixed rule such as:

"If traffic drops by 20%, refresh the page"

is too simple.

Whether a page needs refreshing depends on many factors working together, including search performance, content quality, engagement signals, and historical patterns.

Machine learning can identify relationships between many features simultaneously, making better recommendations than a single if-statement.

The model supports decision-making by ranking pages that are most likely to benefit from a refresh.
